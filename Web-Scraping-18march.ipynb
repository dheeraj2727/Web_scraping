```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Lets us add the url to the notebook
url = 'https://www.flipkart.com/search?q=tv&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off'
url
```

```python
page = requests.get(url)
print(page)
```

```python
soup = BeautifulSoup(page.text,'lxml')
soup
```

```python
# to get only the text we can use a for loop
```

```python
product_name = soup.find_all('div',{'class':"_4rR01T"})
prod_name = []
for i in product_name:
    name = i.text
    prod_name.append(name)
```

```python
prod_name
```

```python
product_price = soup.find_all('div',{'class':"_30jeq3 _1_WHN1"})
prod_price = []
for i in product_price:
    price = i.text
    prod_price.append(price)
    
```

```python
prod_price
```

```python
product_rating = soup.find_all('div',{'class':"_3LWZlK"})
prod_rating = []
for i in product_rating:
    rating = i.text
    prod_rating.append(rating)
```

```python
prod_rating
```
